```python
import itertools
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from table import get_results_folder
from pathlib import Path
from matplotlib.lines import Line2D
from matplotlib.patches import Patch

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
plt.style.use("seaborn-colorblind")
markers = list(Line2D.markers.keys())
prop_cycle = plt.rcParams["axes.prop_cycle"]
default_colours = prop_cycle.by_key()["color"]
hatches = ["", "/", "\\", "|", "-", "x", "o", "+", "O", ".", "*"]
default_hatches = [
    "".join(combi) for combi in list(itertools.product(hatches, hatches))[1:]
]

temp2colour = {"af":default_colours[0], "comb":default_colours[1], "esm":default_colours[2],"none":default_colours[3], "omega":default_colours[4], "truth":default_colours[5]}
msa2marker = {"single":"o", "full":"*"}
shorten = {"af":"AF", "comb":"Comb", "esm":"ESM", "none":"None","omega":"Omega", "truth":"PDB"}
alpha=0.5

orphans = get_results_folder(Path("orphan_data"))
orphans.to_csv("tables/orphans.csv")
orphans
```

```python
temp2colour
```

```python
grouped_orphans=orphans.groupby(["tool", "msa_mode", "template_id"], dropna=False).agg(["mean", "var"])
grouped_orphans
```

```python
[(colour, shorten[identity], alpha) for identity, colour  in temp2colour.items()]
```

```python
matplotlib.rcParams.update({'font.size': 25})
matplotlib.rcParams['figure.figsize'] = (10, 7)
for name, row in grouped_orphans.iterrows():
    temp_id = name[2] if not(str(name[2]) =="nan") else temp_id
    msa_mode = name[1] if not(str(name[1])=="nan") else "single"
    edgecolors= "k" if name[0]!="ColabFold" else "face"
    plt.scatter(row["lddt"]["mean"], row["tm_score"]["mean"], s=np.sqrt(row["tm_score"]["var"]*1000000), edgecolors=edgecolors, c=temp2colour[temp_id], marker=msa2marker[msa_mode], alpha=alpha)
plt.ylabel("lDDT")
plt.xlabel("TM-score")
marker_elements = [Line2D([0],[0], linestyle='', color="k", marker=marker,markersize=20, label=name.capitalize()) for name, marker in msa2marker.items()]
colour_elements = [Patch(facecolor=colour, label=shorten[identity], alpha=alpha) for identity, colour  in temp2colour.items()] 
other_elements = [Line2D([0],[0], linestyle='',markersize=20, markeredgecolor='k', color="w",  marker="s", label="No AF")]
plt.legend(handles=marker_elements+colour_elements+other_elements)
plt.gca().spines["bottom"].set_linewidth(3)
plt.gca().spines["left"].set_linewidth(3)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.title("OmegaFold orphan proteins N=14")
plt.savefig('figures/monomer_orphans.png',bbox_inches="tight", transparent=True)
plt.savefig('figures/monomer_orphans.svg',bbox_inches="tight", transparent=True)
plt.plot()
```

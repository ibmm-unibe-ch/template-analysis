```python
import pandas as pd
from IPython.display import display, HTML
pd.set_option('display.max_rows', 500)
```

```python
orphans = pd.read_csv("orphan_data/table.csv", usecols=["identifier","tool","msa_mode", "template_id","rmsd","tm_score","lddt","plddt","max_pae", "ptm"])[["identifier","tool","msa_mode", "template_id","rmsd","tm_score","lddt","plddt","max_pae", "ptm"]]
sorted = orphans.sort_values(["identifier", "tool", "msa_mode", "template_id"])
```

```python
for identity in sorted.identifier.unique():
    display(HTML(sorted[sorted.identifier==identity].to_html()))
```

```python
import pandas as pd
from IPython.display import display, HTML
pd.set_option('display.max_rows', 500)
```

```python
orphans = pd.read_csv("casp_oligomers/table_multimer.csv", usecols=["identifier","tool", "parts","dockq_scores", "rmsd","tm_score","lddt","plddt","max_pae", "ptm"])[["identifier","tool","parts", "lddt","dockq_scores", "rmsd","tm_score","plddt","max_pae", "ptm"]]
sorted = orphans.sort_values(["identifier", "tool", "parts"])
```

```python
import itertools
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from multimer_table import get_results_folder
from pathlib import Path
from matplotlib.lines import Line2D
import os
matplotlib.rcParams.update({'font.size': 25})
matplotlib.rcParams['figure.figsize'] = (10, 7)

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
plt.style.use("seaborn-colorblind")
prop_cycle = plt.rcParams["axes.prop_cycle"]
default_colours = prop_cycle.by_key()["color"]

hatches = ["", "/", "\\", "|", "-", "x", "o", "+", "O", ".", "*"]
default_hatches = [
    "".join(combi) for combi in list(itertools.product(hatches, hatches))[1:]
]

part2colour = {"af":default_colours[0], "comb":default_colours[1], "esm":default_colours[2],"omega":default_colours[4], "other":default_colours[-1], "truth":default_colours[5]}
part2marker = {"ESM":">","msaaf":"x","fullaf":"*", }
part2marker_name = {"ESM":"ESM","msaaf":"MSA", "fullaf":"Template", }
shorten = {"af":"AF", "comb":"Comb", "esm":"ESM", "msaaf":"MSA", "None":"None","truth":"PDB", "omega":"Omega","other":"Other", "fullaf":"Template", "ColabFold":"AlphaFold"}
alpha=0.5

df = get_results_folder(Path("casp_oligomers"))
os.makedirs("tables", exist_ok=True)
df.to_csv("tables/multimers.csv")
df
```

```python
esm_predictions = set(df[(df.rmsd>0) & (df.tool=="ESM") ].identifier)
esm_groups=df[df.identifier.isin(esm_predictions)].groupby(["tool", "parts"], dropna=False).agg(["mean", "var"])
esm_groups
```

```python
#msaaf -> normal AF
#fullaf -> single sequence, one run
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
import numpy as np

for name, row in esm_groups.iterrows():
    tool, part = name
    colour = part2colour.get(part, part2colour["other"])
    marker = ">" if tool=="ESM" else part2marker.get(part,"o")
    plt.scatter(row["lddt"]["mean"], row["tm_score"]["mean"], s=np.sqrt(row["tm_score"]["var"]*1000000), c=colour, marker=marker, alpha=alpha)
plt.ylabel("lDDT")
plt.xlabel("TM-score")
marker_elements = [Line2D([0],[0], linestyle='', color="k", marker=marker, markersize=20, label=part2marker_name[name]) for name, marker in part2marker.items()  ]
colour_elements = [Patch(facecolor=colour, label=shorten[name],alpha=alpha) for name, colour in part2colour.items()] 
plt.legend(handles=marker_elements+colour_elements)
plt.title("Multimer refinement N=6")
plt.gca().spines["left"].set_linewidth(3)
plt.gca().spines["bottom"].set_linewidth(3)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.savefig('figures/multimer_esm.png',bbox_inches="tight", transparent=True)
plt.savefig('figures/multimer_esm.svg',bbox_inches="tight", transparent=True)
plt.plot()
```
